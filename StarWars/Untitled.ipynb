```python
from StarWars.decisiontree import DecisionTree
from StarWars.scientistvalue import HubbleValue
from StarWars.data import get_data, load_image
from StarWars.data import load_images
import numpy as np
import pandas as pd
from PIL import Image
from tensorflow import keras
```

```python
df = get_data('train', 60000)
```

```python
sample = df.sample(1)
sample_name = np.array(sample['image'])[0]
prediction = np.array(sample)[0]
tree = DecisionTree(prediction) 

result = tree.cat1()

classe = HubbleValue(result)

HubbleValue.final(classe)
```
